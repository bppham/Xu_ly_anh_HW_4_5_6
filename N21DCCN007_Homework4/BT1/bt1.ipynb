```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Hàm hiển thị ảnh
def display_image(data, title, cmap='gray', vmin=None, vmax=None):
    plt.figure()
    plt.imshow(data, cmap=cmap, vmin=vmin, vmax=vmax)
    plt.title(title)
    plt.colorbar()
    plt.show()

# Hàm lưu ảnh
def save_image(data, filename, cmap='gray', vmin=None, vmax=None):
    plt.figure()
    plt.imshow(data, cmap=cmap, vmin=vmin, vmax=vmax)
    plt.axis('off')  # Tắt các trục
    plt.savefig(filename, bbox_inches='tight', pad_inches=0)
    plt.close()  # Đảm bảo không giữ các hình vẽ mở
```

```python
# Bước 1: Khởi tạo các hằng số và thông số
size = 8  # Kích thước ảnh (8x8)
TwoPiOnSize = 2.0 * np.pi / size  # 2 * pi / size
u0 = v0 = 2.0  # Tần số ngang và dọc
```

```python
# Bước 2: Tạo các mảng cho phần thực và ảo
I1R = np.zeros((size, size), dtype=float)  # Phần thực của I1
I1I = np.zeros((size, size), dtype=float)  # Phần ảo của I1
```

```python
# Bước 3: Tính toán các giá trị pixel
for row in range(size):
    for col in range(size):
        Phase = TwoPiOnSize * (u0 * col + v0 * row)
        I1R[row, col] = 0.5 * np.cos(Phase)  # Phần thực
        I1I[row, col] = 0.5 * np.sin(Phase)  # Phần ảo
```

```python
# Bước 4: Hiển thị ảnh phần thực và phần ảo của I1
display_image(I1R, 'Real Part of I1', cmap='gray', vmin=-0.5, vmax=0.5)
display_image(I1I, 'Imaginary Part of I1', cmap='gray', vmin=-0.5, vmax=0.5)

# Lưu ảnh phần thực và phần ảo của I1
save_image(I1R, 'real_I1.png', cmap='gray', vmin=-0.5, vmax=0.5)
save_image(I1I, 'imag_I1.png', cmap='gray', vmin=-0.5, vmax=0.5)
```

```python
# Bước 5: Tính FFT 2D
DFT_I1 = np.fft.fftshift(np.fft.fft2(I1R + 1j * I1I))
```

```python
# Tách phần thực và phần ảo của FFT
real_DFT_I1 = np.real(DFT_I1)
imag_DFT_I1 = np.imag(DFT_I1)
```

```python
# Hiển thị FFT
display_image(real_DFT_I1, 'Real Part of DFT(I1)', cmap='gray')
display_image(imag_DFT_I1, 'Imaginary Part of DFT(I1)', cmap='gray')

# Lưu ảnh FFT phần thực và phần ảo của DFT(I1)
save_image(real_DFT_I1, 'real_DFT_I1.png', cmap='gray')
save_image(imag_DFT_I1, 'imag_DFT_I1.png', cmap='gray')
```

```python
# Bước 6: In ra các giá trị ASCII
print("Re[DFT(I1)]:")
print(np.round(real_DFT_I1, 4))

print("\nIm[DFT(I1)]:")
print(np.round(imag_DFT_I1, 4))
```
